# Mendapatkan plot grafik garis

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("input/Liga120192021.csv")
df.plot()
plt.show()
```

# Pengklasifikasian Data Pada Kolom Pass1 Terhadap Kolom Pass 2

```python
models = {}
for col in df.columns[1:]:
    x = df['Pass1'].values.reshape(-1, 1)
    y = df['Pass2'].values
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
    model = LogisticRegression(max_iter=5000).fit(x_train, y_train)
    models[col] = model
    print("Accuracy for", col, ":", model.score(x_test, y_test))
```

# Pengklasifikasian Data Pada Kolom Pass9 Terhadap Kolom Pass 10

```python
models = {}
for col in df.columns[1:]:
    x = df['Pass9'].values.reshape(-1, 1)
    y = df['Pass10'].values
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
    model = LogisticRegression(max_iter=5000).fit(x_train, y_train)
    models[col] = model
    print("Accuracy for", col, ":", model.score(x_test, y_test))
```

# Confusion Matrix pada Pass 1

```python
x = df.iloc[:, 1:].values
y = df['Pass1'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
model = LogisticRegression(max_iter=10000).fit(x_train, y_train)
y_pred = model.predict(x_test)
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

# Confusion Matrix pada Pass 9

```python
x = df.iloc[:, 1:].values
y = df['Pass9'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
model = LogisticRegression(max_iter=10000).fit(x_train, y_train)
y_pred = model.predict(x_test)
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

# Cluster 2 model kolom

```python
x = df.iloc[:, 1:].values
kmeans = KMeans(n_clusters=2, random_state=0).fit(x)
print("\nCluster Centers:")
print(kmeans.cluster_centers_)
```

```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
x = df.iloc[:, :-1].values
k = 2
agg_clustering = AgglomerativeClustering(n_clusters=k)
y_pred = agg_clustering.fit_predict(x)
colors = ['red', 'blue']
for i in range(k):
    plt.scatter(x[y_pred == i, 0], x[y_pred == i, 1], s=30, c=colors[i], label='Cluster' + str(i+1))
plt.legend()
plt.show()
```
